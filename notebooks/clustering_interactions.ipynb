```python
import pandas as pd
import ast
import math
```

### Collapsed similar interactions to get common sequences

```python
data = pd.read_csv('./data/collapsed_interactions_data.csv')
data_null = pd.read_csv('./data/dataset_for_task_level_without_interactions.csv')
```

```python
data['gotzzhou2009_interactions'] = data.apply(lambda row: ast.literal_eval(row['gotzzhou2009_interactions']), axis=1)
data['gotzzhou2009_guo2015'] = data.apply(lambda row: ast.literal_eval(row['gotzzhou2009_guo2015']), axis=1)
data['gotzzhou2009_shneiderman1996'] = data.apply(lambda row: ast.literal_eval(row['gotzzhou2009_shneiderman1996']), axis=1)
data['gotzzhou2009_gotzwen2009'] = data.apply(lambda row: ast.literal_eval(row['gotzzhou2009_gotzwen2009']), axis=1)

data_null['gotzzhou2009_interactions'] = data_null.apply(lambda row: ast.literal_eval(row['gotzzhou2009_interactions']), axis=1)
data_null['gotzzhou2009_guo2015'] = data_null.apply(lambda row: ast.literal_eval(row['gotzzhou2009_guo2015']), axis=1)
data_null['gotzzhou2009_shneiderman1996'] = data_null.apply(lambda row: ast.literal_eval(row['gotzzhou2009_shneiderman1996']), axis=1)
data_null['gotzzhou2009_gotzwen2009'] = data_null.apply(lambda row: ast.literal_eval(row['gotzzhou2009_gotzwen2009']), axis=1)
```

```python
data['gotzzhou2009_interactions_len'] = data.apply(lambda row: len(row['gotzzhou2009_interactions']), axis=1)
data['gotzzhou2009_guo2015_len'] = data.apply(lambda row: len(row['gotzzhou2009_guo2015']), axis=1)
data['gotzzhou2009_shneiderman1996_len'] = data.apply(lambda row: len(row['gotzzhou2009_shneiderman1996']), axis=1)
data['gotzzhou2009_gotzwen2009_len'] = data.apply(lambda row: len(row['gotzzhou2009_gotzwen2009']), axis=1)
```

```python
data['ddt'] = data.apply(lambda row: f'{row.dataset}_{row.internal_dataset}_{row.task}', axis=1)
data['ddtu'] = data.apply(lambda row: f'{row.dataset}_{row.internal_dataset}_{row.task}_{row.user_id}', axis=1)
```

```python
data
```

```python
cols = ['userstudy_data_task', 'interaction_mean','guo2015_seq_mean', 'shneiderman1996_seq_mean', 'gotzwen2009_seq_mean']
mean_counts = pd.DataFrame(columns=cols)
#mean_counts.set_index('userstudy_data_task')
for i in data.ddt.unique():
    temp = data[data.ddt == i]
    mean_counts = mean_counts.append(pd.Series([i,temp.gotzzhou2009_interactions_len.mean() ,temp.gotzzhou2009_guo2015_len.mean(),temp.gotzzhou2009_shneiderman1996_len.mean(),temp.gotzzhou2009_gotzwen2009_len.mean()], index=cols), ignore_index=True)
```

```python
mean_counts
```

```python
dataset = 'BattleHeer2019'
freq_counts = {}
lower_bound = 3
upper_bound = 15

data_subset = data[data.dataset == dataset]
print(len(data_subset))

for temp in data_subset['gotzzhou2009_interactions']:
    for i in range(len(temp)):
        for j in range(len(temp)+1):
            ss = tuple(temp[i:j])
            if len(ss) >= lower_bound and len(ss) <= upper_bound:
                if ss in freq_counts.keys():
                    freq_counts[ss] += 1
                else:
                    freq_counts[ss] = 1
```

```python
dict(sorted(freq_counts.items(), key=lambda item: item[1], reverse=True))
```
